### Import the Data

```python
# required libraries and dependencies
import pandas as pd
import numpy as np
from pathlib import Path
%matplotlib inline
```

```python
# Import the data by reading in the CSV file and setting the DatetimeIndex 

analyzing_portfolios_df = pd.read_csv(
    Path("Resources/whale_navs.csv"),
    index_col=("date"),
    parse_dates=True,
    infer_datetime_format=True)
    
analyzing_portfolios_df.head()
```

```python
# Prepare for the analysis by converting the dataframe of NAVs and prices to daily returns
# Drop any rows with all missing values
# Review the daily returns DataFrame.
daily_returns_df = analyzing_portfolios_df.pct_change().dropna()
daily_returns_df.head()

```

### Analyze the Risk

Evaluate the risk profile of each portfolio by using the standard deviation and the beta.

```python
# Calculate and sort the standard deviation for all 4 portfolios and the S&P 500
# Review the standard deviations sorted smallest to largest
std_dev = daily_returns_df.std().sort_values()
std_dev
```

```python
# Calculate and sort the annualized standard deviation (252 trading days) of the 4 portfolios and the S&P 500
# Review the annual standard deviations smallest to largest
annualized_std_dev = std_dev * np.sqrt(252)
annualized_std_dev.sort_values()
```

```python
# Using the daily returns DataFrame and a 21-day rolling window, 
# plot the rolling standard deviation of the 4 portfolios and the S&P 500
# Include a title parameter and adjust the figure size
rolling_std_dev = daily_returns_df.rolling(window = 21).std()
rolling_std_dev.plot(figsize=(14, 10), title="21 days Rolling standard Deviation" )
```

```python
# Using the daily return data and a 21-day rolling window, plot the rolling standard deviation of just the 4 portfolios. 
# Include a title parameter and adjust the figure size
four_funds_rolling_std_dev = daily_returns_df.rolling(window = 21).std()
rolling_std_dev.drop(["S&P 500"], axis=1).plot(figsize=(14, 10), title="4 funds 21 days Rolling standard Deviation" )
```

### Analyze the Risk-Return Profile



```python
# Calculate the annual average return data for the for fund portfolios and the S&P 500
# Use 252 as the number of trading days in the year
# Review the annual average returns sorted from lowest to highest
year_trading_days=252
annual_average_return = daily_returns_df.mean() * year_trading_days
annual_average_return.sort_values()
```

```python
# Calculate the annualized Sharpe Ratios for each of the 4 portfolios and the S&P 500.
# Review the Sharpe ratios sorted lowest to highest
annualized_sharpe_ratios = annual_average_return / annualized_std_dev
annualized_sharpe_ratios.sort_values()
```

```python
# Visualize the Sharpe ratios as a bar chart
# Include a title parameter and adjust the figure size
annualized_sharpe_ratios.plot.bar(title="annualized sharpe ratios")
```

### Diversify the Portfolio



```python
# Calculate the variance of the S&P 500 using a rolling 60-day window.
market_variance = daily_returns_df["S&P 500"].rolling(window = 60).var()
market_variance.tail()
```

##### Portfolio 1 

```python
# Calculate the covariance using a 60-day rolling window 
# Review the last five rows of the covariance data
berkshire_covariance = daily_returns_df["BERKSHIRE HATHAWAY INC"].rolling(window = 60).cov(daily_returns_df["S&P 500"])
berkshire_covariance.tail()
```

##### Portfolio 1 

```python
# Calculate the beta based on the 60-day rolling covariance compared to the market (S&P 500)
# Review the last five rows of the beta information
berkshire_beta= berkshire_covariance /market_variance
berkshire_beta.tail()
```

##### Portfolio 1 

```python
# Calculate the average of the 60-day rolling beta
berkshire_beta.mean()
```

##### Portfolio 1 

```python
# Plot the rolling beta 
# Include a title parameter and adjust the figure size


berkshire_beta.plot(title="berkshire 60 days Rolling beta")
```

##### Portfolio 2 

```python
# Calculate the covariance using a 60-day rolling window 
# Review the last five rows of the covariance data
tiger_covariance = daily_returns_df["TIGER GLOBAL MANAGEMENT LLC"].rolling(window = 60).cov(daily_returns_df["S&P 500"])
tiger_covariance.tail()
```

##### Portfolio 2 

```python
# Calculate the beta based on the 60-day rolling covariance compared to the market (S&P 500)
# Review the last five rows of the beta information
tiger_beta = tiger_covariance / market_variance
tiger_beta.tail()
```

##### Portfolio 2 

```python
# Calculate the average of the 60-day rolling beta
tiger_beta.mean()
```

##### Portfolio 2 

```python
# Plot the rolling beta 
# Include a title parameter and adjust the figure size

tiger_beta.plot(title="tiger 60 days Rolling beta")
```
